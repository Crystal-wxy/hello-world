```python
import pandas as pd
```

```python
#data = pd.read_csv('shuffle.csv', sep='\t')
data = pd.read_csv('server_shuffle_wjj.csv', sep='  ')
```

```python
data
```

```python
import numpy as np
```

```python
log_data = data.apply(np.log2, axis=1)
```

```python
log_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
```

```python
plt.figure()
plt.plot(data['Size (bytes)'], data['Runtime (sec)'])
```

```python
params = {
            'axes.labelsize': '16',
            'xtick.labelsize': '16',
            'ytick.labelsize': '14',
            'lines.linewidth': '2',
            'legend.fontsize': '20',
            #'title.fontsize': '20',
        }
pylab.rcParams.update(params)
```

```python
fig = plt.figure(figsize=(12,8))
fig.suptitle('shuffled data access on server', fontsize=30)
plt.plot(log_data['Size (bytes)'], log_data['Runtime (sec)'], '-o')

l1_idx = 11
l1_log_x = log_data['Size (bytes)'][l1_idx]
l1_log_y = log_data['Runtime (sec)'][l1_idx]

l2_idx = 23
l2_log_x = (log_data['Size (bytes)'][l2_idx] + log_data['Size (bytes)'][l2_idx + 1]) / 2
l2_log_y = (log_data['Runtime (sec)'][l2_idx] + log_data['Runtime (sec)'][l2_idx + 1]) / 2

l3_idx = 49
l3_log_x = log_data['Size (bytes)'][l3_idx]
l3_log_y = log_data['Runtime (sec)'][l3_idx]

real_use_l2_idx = 19
real_use_l2_x = log_data['Size (bytes)'][real_use_l2_idx]
real_use_l2_y = log_data['Runtime (sec)'][real_use_l2_idx]

plt.plot(l1_log_x, l1_log_y, 'ro')
show_l1='32KIB(L1d)'
plt.annotate(show_l1, xytext=(l1_log_x + 0.2, l1_log_y - 0.05), xy=(l1_log_x, l1_log_y), fontsize=16)

plt.plot(l2_log_x, l2_log_y, 'ro')
show_l2 = '256KIB(L2)'
plt.annotate(show_l2, xytext=(l2_log_x + 0.3, l2_log_y - 0.15), xy=(l2_log_x, l2_log_y), fontsize=16)

plt.plot(l3_log_x, l3_log_y, 'ro')
show_l3 = '25600KIB(L3)'

plt.annotate(show_l3, xytext=(l3_log_x + 0.3, l3_log_y - 0.1), xy=(l3_log_x, l3_log_y), fontsize=16)
plt.xlabel('log(Size (bytes))')
plt.ylabel('log(Runtime (seconds))')
```
